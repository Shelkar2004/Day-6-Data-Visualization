```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame with columns 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', and 'Species'
df = pd.DataFrame({
    'SepalLengthCm': [5.1, 4.9, 4.7, 4.6, 5.0, 5.4],
    'SepalWidthCm': [3.5, 3.0, 3.2, 3.1, 3.6, 3.9],
    'PetalLengthCm': [1.4, 1.3, 1.5, 1.4, 1.7, 1.4],
    'PetalWidthCm': [0.2, 0.2, 0.2, 0.2, 0.4, 0.3],
    'Species': ['A', 'B', 'A', 'C', 'B', 'A']
})

# Define colors for each species
colors = {'A': 'red', 'B': 'blue', 'C': 'green'}

# Get the number of features
num_features = len(df.columns) - 1

# Create a pair plot
fig, axes = plt.subplots(num_features, num_features, figsize=(12, 12))

for i, feature1 in enumerate(df.columns[:-1]):
    axes[i, i].hist(df[feature1], bins=10, color='gray', alpha=0.7)
    axes[i, i].set_title(feature1)

    for j, feature2 in enumerate(df.columns[:-1]):
        if i != j:
            for species, color in colors.items():
                axes[i, j].scatter(df.loc[df['Species'] == species, feature2], df.loc[df['Species'] == species, feature1], label=species, color=color, alpha=0.5)

# Adjust layout
plt.tight_layout()

# Add legend to the last subplot
axes[0, num_features - 1].legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming df is your DataFrame with columns 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', and 'Species'
df = pd.DataFrame({
    'SepalLengthCm': [5.1, 4.9, 4.7, 4.6, 5.0, 5.4],
    'SepalWidthCm': [3.5, 3.0, 3.2, 3.1, 3.6, 3.9],
    'PetalLengthCm': [1.4, 1.3, 1.5, 1.4, 1.7, 1.4],
    'PetalWidthCm': [0.2, 0.2, 0.2, 0.2, 0.4, 0.3],
    'Species': ['A', 'B', 'A', 'C', 'B', 'A']
})

# Define colors for each species
colors = {'A': 'red', 'B': 'blue', 'C': 'green'}

# Create a set of distribution plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for i, feature in enumerate(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']):
    for species, color in colors.items():
        subset = df[df['Species'] == species]
        axes.flatten()[i].hist(subset[feature], bins=10, alpha=0.5, label=species, color=color)

    axes.flatten()[i].set_title(feature)
    axes.flatten()[i].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python

# Assuming df is your DataFrame with numeric columns
df = pd.DataFrame({
    'SepalLengthCm': [5.1, 4.9, 4.7, 4.6, 5.0, 5.4],
    'SepalWidthCm': [3.5, 3.0, 3.2, 3.1, 3.6, 3.9],
    'PetalLengthCm': [1.4, 1.3, 1.5, 1.4, 1.7, 1.4],
    'PetalWidthCm': [0.2, 0.2, 0.2, 0.2, 0.4, 0.3],
})

# Calculate correlation matrix
correlation_matrix = df.corr(method='pearson')

# Create a heatmap using plt.matshow
plt.matshow(correlation_matrix, cmap='coolwarm')

# Add annotations
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.index)):
        plt.text(i, j, f'{correlation_matrix.iloc[j, i]:.2f}', ha='center', va='center', color='black')

# Add colorbar
plt.colorbar()

# Set axis labels
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
plt.yticks(range(len(correlation_matrix.index)), correlation_matrix.index)

# Show the plot
plt.show()
```

```python
# Assuming df is your DataFrame with columns 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm', and 'Species'
df = pd.DataFrame({
    'SepalLengthCm': [5.1, 4.9, 4.7, 4.6, 5.0, 5.4],
    'SepalWidthCm': [3.5, 3.0, 3.2, 3.1, 3.6, 3.9],
    'PetalLengthCm': [1.4, 1.3, 1.5, 1.4, 1.7, 1.4],
    'PetalWidthCm': [0.2, 0.2, 0.2, 0.2, 0.4, 0.3],
    'Species': ['A', 'B', 'A', 'C', 'B', 'A']
})

# Define colors for each species
colors = {'A': 'red', 'B': 'blue', 'C': 'green'}

# Create boxplots using Matplotlib
plt.figure(figsize=(10, 10))

# Boxplot for SepalLengthCm
plt.subplot(221)
for species, color in colors.items():
    subset = df[df['Species'] == species]
    plt.boxplot(subset['SepalLengthCm'], labels=[species], showfliers=False, patch_artist=True, boxprops=dict(facecolor=color))

# Boxplot for SepalWidthCm
plt.subplot(222)
for species, color in colors.items():
    subset = df[df['Species'] == species]
    plt.boxplot(subset['SepalWidthCm'], labels=[species], showfliers=False, patch_artist=True, boxprops=dict(facecolor=color))

# Boxplot for PetalLengthCm
plt.subplot(223)
for species, color in colors.items():
    subset = df[df['Species'] == species]
    plt.boxplot(subset['PetalLengthCm'], labels=[species], showfliers=False, patch_artist=True, boxprops=dict(facecolor=color))

# Boxplot for PetalWidthCm
plt.subplot(224)
for species, color in colors.items():
    subset = df[df['Species'] == species]
    plt.boxplot(subset['PetalWidthCm'], labels=[species], showfliers=False, patch_artist=True, boxprops=dict(facecolor=color))

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Check if 'Id' column exists before dropping it
if 'Id' in df.columns:
    # Drop the 'Id' column for correlation calculation
    correlation_matrix = df.drop('Id', axis=1).corr()
else:
    # If 'Id' column doesn't exist, drop nothing
    correlation_matrix = df.corr()

# Get the column names for labels
features = correlation_matrix.columns

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Plot the heatmap
cax = ax.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')

# Set labels and title
ax.set_xticks(np.arange(len(features)))
ax.set_yticks(np.arange(len(features)))
ax.set_xticklabels(features)
ax.set_yticklabels(features)
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Add colorbar
cbar = fig.colorbar(cax, shrink=0.8, aspect=5)
cbar.set_label('Correlation')

# Show the plot
plt.show()
```
